```python
import cv2
from matplotlib import pyplot as plt

# Load the two images
img1 = cv2.imread('girafe_coarse.png')
img2 = cv2.imread('uno.png')

# Resize the images to the same size
img1 = cv2.resize(img1, (500, 500))
img2 = cv2.resize(img2, (500, 500))

# Merge the images
merged_img = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)

# Display the merged image
#cv2.imshow('Merged Image', merged_img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

RGB_im = cv2.cvtColor(merged_img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_im)
plt.title('Merged Image')
plt.show()
```

```python
import cv2
import numpy as np

# Load the two images
img1 = cv2.imread('girafe_coarse.png')
img2 = cv2.imread('uno.png')

# Resize the images to the same height
height = 500
width1 = int(img1.shape[1] * height / img1.shape[0])
width2 = int(img2.shape[1] * height / img2.shape[0])
img1 = cv2.resize(img1, (width1, height))
img2 = cv2.resize(img2, (width2, height))

# Create a black canvas to place the images side by side
canvas = np.zeros((height, width1 + width2, 3), dtype=np.uint8)

# Place the first image on the left side of the canvas
canvas[:, :width1, :] = img1

# Place the second image on the right side of the canvas
canvas[:, width1:, :] = img2

# Display the merged image
cv2.imshow('Merged Image', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the two images
img1 = cv2.imread('girafe_coarse.png')
img2 = cv2.imread('uno.png')

# Resize the images to the same width
width = 500
height1 = int(img1.shape[0] * width / img1.shape[1])
height2 = int(img2.shape[0] * width / img2.shape[1])
img1 = cv2.resize(img1, (width, height1))
img2 = cv2.resize(img2, (width, height2))

# Create a black canvas to place the images vertically
canvas = np.zeros((height1 + height2, width, 3), dtype=np.uint8)

# Place the first image on the top of the canvas
canvas[:height1, :, :] = img1

# Place the second image on the bottom of the canvas
canvas[height1:, :, :] = img2

# Display the merged image
cv2.imshow('Merged Image', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
